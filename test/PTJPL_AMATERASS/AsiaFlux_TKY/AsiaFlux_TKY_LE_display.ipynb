### Latent Heat Flux
Latent heat flux is generally referred to as evapotranspiration, which describes the total evaporation from land surface and transpiration by plants.

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ec_data_file = '20180719.csv'
```

```python
ec_MMS = pd.read_csv(ec_data_file) # 2018 200day <-> 2018.07.19
ec0719_MMS = ec_MMS.set_index('TIME')
```

```python
ec0719_MMS
```

```python
hour_list = [(4-len(str(hour_str)))*'0'+str(hour_str)[:-2]+':'+str(hour_str)[-2:] for hour_str in ec0719_MMS.index]
LE_list = numpy.array(ec0719_MMS.LE_spec)
```

```python
plt.figure()
plt.plot(LE_list, label='the latent heat flux')
plt.xticks(ticks=numpy.arange(len(hour_list)), labels=hour_list, rotation=90, ha='right')
plt.xlim(0,len(hour_list)-1)
# plt.ylim(0, 300)
plt.legend()
plt.ylabel('$Wm^{-2}$', fontsize=14)
plt.title('Takayama Deciduous Broadleaf Forest 2018/07/19')
plt.grid(True)
```
