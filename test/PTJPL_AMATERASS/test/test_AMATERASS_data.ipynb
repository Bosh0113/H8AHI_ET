```python
import numpy
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.ticker as tck
```

## Downward SW flux at the surface [W/m2]

```python
data_path = r'D:\PhD_Workspace\H8_ET\231016\JP\201808010130.dwn.sw.flx.sfc.msm.1km.bin'
dwn_dn = numpy.fromfile(data_path, dtype='>f4').reshape(2521, 3001)

jp_extent = (120.0, 150.0, 22.4, 47.6)
proj=ccrs.PlateCarree()
ax=plt.axes(projection=proj)
ax.set_extent(jp_extent,proj)
ax.coastlines(resolution='10m')
gl=ax.gridlines(crs=proj,draw_labels=True)
gl.xlabels_top = False
gl.ylabels_right = False
gl.xlocator = tck.FixedLocator(numpy.arange(120,150,5))
gl.ylocator = tck.FixedLocator(numpy.arange(20,50,5))
plt.imshow(dwn_dn, extent=jp_extent, cmap='gist_ncar', vmin=0.0, vmax=1400.0, interpolation='None')
plt.colorbar()
```
