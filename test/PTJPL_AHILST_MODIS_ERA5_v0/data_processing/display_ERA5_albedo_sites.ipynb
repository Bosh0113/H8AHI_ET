```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
def read_site_albedo(csv_filename):
    csv_data = numpy.genfromtxt(csv_filename, delimiter=',', dtype=str)
    albedo_hour = csv_data[:, 1]
    albedo_day = albedo_hour.reshape(365*2, 24)
    albedo_day = albedo_day.astype(float)
    return albedo_day
```

```python
ERA5_ALBEDO_SITE_CSV = '/disk2/Data/FluxSite4ET/TKY_ERA5_Albedo.csv'
albedo_day = read_site_albedo(ERA5_ALBEDO_SITE_CSV)
plt.figure(figsize=(10,4))
plt.plot(numpy.arange(365*2), albedo_day, '.', color='green')
plt.title('TKY 2018-2019')
plt.ylabel('ERA5 Albedo')
```
