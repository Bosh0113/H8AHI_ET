```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
def read_site_ref(csv_filename):
    csv_data = numpy.genfromtxt(csv_filename, delimiter=',', dtype=str)
    res_day1 = csv_data[:, 1]
    res_day1 = res_day1.astype(float)
    res_day2 = csv_data[:, 2]
    res_day2 = res_day2.astype(float)
    return res_day1, res_day2
```

```python
site_names = ['TKY','CRK', 'GCK', 'GDK']

for site_name in site_names:
    MOD43GA_REF_SITE_CSV = '/disk2/Data/FluxSite4ET/original/' + site_name + '_MOD09GA_ReflectanceB1B2.csv'
    ref1_day, ref2_day = read_site_ref(MOD43GA_REF_SITE_CSV)
    plt.figure(figsize=(10,4))
    plt.plot(numpy.arange(365*2), ref1_day, color='red')
    plt.plot(numpy.arange(365*2), ref2_day, color='black')
    plt.title(site_name + ' 2018-2019')
    plt.xlim(0, 365*2)
    plt.ylabel('MOD43GA Reflectance')
    plt.show()
```
