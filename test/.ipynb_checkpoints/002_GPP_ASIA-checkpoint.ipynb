```python
!pip install cartopy
import cartopy
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.ticker as mticker
import pandas as pd
import xarray as xr
import cartopy.crs as ccrs
import cartopy.util as cutil
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
from matplotlib import colors
from matplotlib.colors import ListedColormap ,BoundaryNorm
from matplotlib.cm import ScalarMappable, get_cmap
import netCDF4
from netCDF4 import Dataset
import scipy.optimize
from scipy import linalg
from scipy import stats as st
from scipy.stats import pearsonr
from scipy import signal
import tifffile
import datetime
from sklearn.metrics import mean_squared_error
import random
import seaborn as sns
from statistics import mean
import openpyxl
from matplotlib.axes import Axes
from PIL import Image
import os
# 기준 경로
base_path = "/home/beak/ドキュメント/GitHub/D1_visit/akane/ALLVAR/"
base_path1 = "/home/beak/ドキュメント/GitHub/D1_visit/akane/"
#dir_savefigure = "/home/beak/beak_data/20231003"
```

```python
# 아시아 지역의 경도와 위도 범위 설정
asia_extent = (60, 180, -10, 80)  # (서경, 동경, 남위, 북위)

#VISIT_2001-2015 mean GPP
startyear=2001
endyear=2015

fill = open(base_path + "ALLVAR_olson_Carbon_14_" + str(1900) + ".mon.bsq.flt", "rb")
fill = np.fromfile(fill, dtype='float32')
fill = np.array(fill.reshape(12, 360, 720))
fill = fill[0, :, :]  # -9999=Sea

databox1 = np.zeros((15, 360, 720), dtype='float32')

for year in range(startyear,endyear+1,1):
    data=open(base_path + "ALLVAR_olson_Carbon_14_"+str(year)+".mon.bsq.flt","rb")
    data=np.fromfile(data,dtype='float32')
    data=np.array(data.reshape(12,360,720),dtype=float)
    data=np.where(data<-9000,np.nan,data)
    databox1[year-startyear,:,:]=np.nansum(data,axis=0)
    
lat_array = np.arange(89.25, -89.75, -0.5)
lon_array = np.arange(-180.25, 179.75, 0.5)

# 아시아 지역에 해당하는 부분 추출

asia_lat_indices = np.where((lat_array >= asia_extent[2]) & (lat_array <= asia_extent[3]))[0]
asia_lon_indices = np.where((lon_array >= asia_extent[0]) & (lon_array <= asia_extent[1]))[0]

result=np.nanmean(databox1,axis=0)*100 #평균GPP계산 NaN무시
result=np.where(fill<-9000,np.nan,result)

# 아시아 지역의 평균 GPP 데이터 추출
asia_gpp_data = result[asia_lat_indices, :][:, asia_lon_indices]
print(asia_gpp_data.shape)# Define the bounding box for the entire world
xmin, xmax, ymin, ymax = -180, 180, -90, 90


# Create the figure and subplot with the correct projection
fig, ax = plt.subplots(figsize=(10, 5), dpi=700, subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent(asia_extent, crs=ccrs.PlateCarree())
ax =plt.subplot(projection=ccrs.PlateCarree())
ax.coastlines(resolution='10m', color='black', linestyle='-', linewidth=0.1, alpha=0.9)
# 경도와 위도 눈금 설정
xstep, ystep = 20, 10
#ax.set_xticks(np.arange(box[0], box[1] + xstep, xstep), crs=ccrs.PlateCarree())
#ax.set_yticks(np.arange(box[2], box[3] + ystep, ystep), crs=ccrs.PlateCarree())
ax.set_xticks(np.arange(xmin, xmax + xstep, xstep), crs=ccrs.PlateCarree())
ax.set_yticks(np.arange(ymin, ymax + ystep, ystep), crs=ccrs.PlateCarree())
# 경도와 위도 눈금 라벨 설정
lon_formatter = LongitudeFormatter(zero_direction_label=False)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)


# 해안선 추가
ax.add_feature(cfeature.COASTLINE, linewidth=0.5, edgecolor='black')

# 공간 분포 맵 시각화
im = plt.imshow(asia_gpp_data, cmap='viridis_r', extent=(lon_array[asia_lon_indices[0]], lon_array[asia_lon_indices[-1]], lat_array[asia_lat_indices[-1]], lat_array[asia_lat_indices[0]]))

# 다른 시각화 설정을 추가할 수 있습니다.
plt.colorbar(im, label='GPP [gC m$^{-2}$ year$^{-1}$]')
plt.title('Average GPP Distribution Map (2001-2015)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# 맵을 표시합니다.
plt.grid(True)
plt.show()
```

```python
# Plot the GPP over the years for Asia
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(range(startyear, endyear + 1), np.nanmean(databox1[:, asia_lat_indices, :][:, :, asia_lon_indices], axis=(1, 2)) * 100, marker='o', linestyle='-', color='r', label='Asia')
ax.plot(range(startyear, endyear + 1), np.nanmean(databox1, axis=(1, 2)) * 100, marker='o', linestyle='-', color='b', label='Global')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('GPP [gC m$^{-2}$ year$^{-1}$]', fontsize=12)
ax.set_title('Average GPP Over the Years (2001-2015)', fontsize=14)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
