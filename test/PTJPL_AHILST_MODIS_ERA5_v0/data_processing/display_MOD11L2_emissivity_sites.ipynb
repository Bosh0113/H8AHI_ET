```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
def read_site_emis(csv_filename):
    csv_data = numpy.genfromtxt(csv_filename, delimiter=',', dtype=str)
    emis_day = csv_data[:, 1]
    emis_day = emis_day.astype(float)
    return emis_day
```

```python
MOD11L2_EMIS_SITE_CSV = '/disk2/Data/FluxSite4ET/TKY_MOD11L2_Emissiviry.csv'
emis_day = read_site_emis(MOD11L2_EMIS_SITE_CSV)
plt.figure(figsize=(10,4))
plt.plot(numpy.arange(365*2), emis_day, '.', color='blue')
plt.title('TKY 2018-2019')
plt.xlim(0, 365*2)
plt.ylabel('MOD11L2 Emissivity')
```
