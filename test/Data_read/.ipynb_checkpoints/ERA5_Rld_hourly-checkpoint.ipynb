```python
# JP extent
resolution = 0.01 # degree
lats = numpy.arange(50.-resolution/2, 20, -resolution)
lons = numpy.arange(120.+resolution/2, 150, resolution)
```

```python
day_num = 1 # 1 -> 31(?)
hour_idx = 2 # 0 <=> 01:00:00

rld_dh = rld_ds.isel(time=day_num, step=hour_idx)
jp_ds = rld_dh.interp(longitude=lons, latitude=lats, method="linear", kwargs={"fill_value": "extrapolate"}) # linear
rld_v = jp_ds.to_array()[0]/3600
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rld_v, interpolation='None')
plt.colorbar(label='Rld')
plt.title('0.01° Rld in study extent at 2018-01-01 03:00:00')
```
