```python
import gzip
import numpy
import matplotlib.pyplot as plt
```

```python
ahi_lst_gz = '/disk2/Data/AHILST_YY/201808/AHILST.v0.201808010300.dat.gz'
with gzip.open(ahi_lst_gz, 'rb') as file:
    gz_data = file.read()
    lst_data = numpy.copy(numpy.frombuffer(gz_data, dtype='f4').reshape(1500, 1500)) # 0.02°
```

```python
lst_data[lst_data==-999]=numpy.NaN
plt.figure(figsize=(10,10))
plt.imshow(lst_data, cmap='gist_ncar', vmin=-20, vmax=50, interpolation='None')
plt.colorbar(label='AHI Land Surface Temperature (℃)')
plt.title('AHI Land Surface Temperature (℃) at UTC 2018-08-01 03:00:00')
```
