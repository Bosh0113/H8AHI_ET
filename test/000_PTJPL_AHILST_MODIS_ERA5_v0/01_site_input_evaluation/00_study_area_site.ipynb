```python
import os
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
site = ['TKY','CRK','GCK','GDK']
lon = [137.423,127.251,127.162,127.149]
lat = [36.146,38.201,37.748,37.749]
```

```python
storage_folder = '/disk2/workspace/20231124'
```

## study area

```python
fig = plt.figure(figsize=(8,7))
# plt.rc('font',family='Times New Roman')
m = Basemap(projection='cyl', resolution='h',
            llcrnrlat=22.5, urcrnrlat = 47.5,
            llcrnrlon=120, urcrnrlon = 150,
            )  
m.shadedrelief()
m.drawparallels(np.arange(20, 50, 5), labels=[1,0,0,0], fontsize=15,linewidth=0.05)
m.drawmeridians(np.arange(120, 151, 5), labels=[0,0,0,1], fontsize=15,linewidth=0.05)

for i in range(len(site)):
    font1 = {'weight':'normal','fontsize':15}
    plt.plot(float(lon[i]), float(lat[i]), marker='o', color='black', markersize=7, markerfacecolor='white')
#     plt.annotate(site[i], xy = (float(lon[i]), float(lat[i])),xytext = (float(lon[i])+0.5, float(lat[i])+0.3), fontsize = 12,color='k',)
```

```python
fig.savefig(os.path.join(storage_folder, 'study_area_JP.png'), dpi=500)
```
