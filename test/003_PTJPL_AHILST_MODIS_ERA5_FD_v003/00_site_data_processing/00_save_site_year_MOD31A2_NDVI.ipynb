```python
import numpy
import os
import xarray
from datetime import datetime, timedelta
from scipy.interpolate import interp1d
```

```python
WORKSPACE_FOLDER = os.path.join(os.path.abspath('..'), 'processing_record')
if not os.path.exists(WORKSPACE_FOLDER):
    os.makedirs(WORKSPACE_FOLDER)
MOD31A2_NDVI_FLODER = '/data01/people/beichen/data_fd_et/MOD31A2_NDVI'
```

```python
def read_ndvi_sites(modis_ndvi_tif, site_infos):
    modis_ds = xarray.open_rasterio(modis_ndvi_tif)[0]
    
    site_v_list = []
    for site_info in site_infos:
        lat_site = site_info[1]
        lon_site = site_info[2]
        site_ds = modis_ds.interp(x=[lon_site], y=[lat_site], method="nearest", kwargs={"fill_value": "extrapolate"}) # linear
        ndvi_dn = numpy.array(site_ds.values)[0][0]
        if ndvi_dn == -3000:
            site_v_list.append(numpy.NaN)
        else:
            data_v = ndvi_dn*0.0001
            site_v_list.append(data_v)
    
    return site_v_list
```

```python
site_infos = [# site, lat, lon, UTC+
    ["CLC", 44.59666667, 123.4702778, 8],
    ["JRC", 31.80683056, 119.2172639, 8],
    ["PDC", 26.6, 106.3166667, 8],
    ["PJC", 40.94138889, 121.9602778, 8],
    ["HZF", 51.78111111, 121.0177778, 8],
    ["ZAG", 32.8, 102.55, 8],
    ["DBB", 41.64397222, 110.3314722, 8],
    ["HBG", 37.61666667, 101.3166667, 8],
    ["MPM", 22.4982, 114.0292, 8],
    ["BBY", 43.323, 141.8107, 9],
    ["TKY", 36.14616667, 137.4231111, 9],
    ["CRK", 38.20138889, 127.2505556, 9],
    ["GCK", 37.74833333, 127.1622222, 9],
    ["GDK", 37.74888889, 127.1491667, 9],
    ["SIP", 0.395204167, 102.7645506, 7],
    ["SDP", 0.699491389, 102.793305, 7],
    ["ASM", -22.2828, 133.2493, 9],
    ["CPW", -33.6152, 150.7236, 11],
    ["WAR", -43.095, 146.6545, 11]
]
```

```python
# 1. record full day from 20180101 to 20191231 in daily
# 2. extend daily into every 10-min 
# 3. save site data

time_internal = 10 # min
START_TIME = '2018-01-01T00:00:00Z' # local time
END_TIME = '2019-12-31T23:59:59Z'

years = ['2018', '2019']
year_days = [(3-len(str(day_n)))*'0' + str(day_n) for day_n in numpy.arange(1, 366, 16)]

data_record = [] # daily
for site_idx in range(len(site_infos)):
    data_record.append([])

for year_str in years:
    for day_idx in range(len(year_days)):
        day_str = year_days[day_idx]
        ndvi_filename = os.path.join(MOD31A2_NDVI_FLODER, 'MOD13A2.061__1_km_16_days_NDVI_doy' + year_str + day_str + '_aid0001.tif')

        day_site_list = read_ndvi_sites(ndvi_filename, site_infos)
        
        for site_idx in range(len(site_infos)):
            record_item = data_record[site_idx]
            day_site_v = day_site_list[site_idx]
            record_item.append(day_site_v)
            data_record[site_idx] = record_item
```

```python
data_first_date = datetime.strptime(START_TIME, "%Y-%m-%dT%H:%M:%SZ")
for site_idx in range(len(site_infos)):
    site_v_d = numpy.array(data_record[site_idx])
    # filldata
    site_v_d = site_v_d
    valid_indices = numpy.where(~numpy.isnan(site_v_d))[0]
    if len(valid_indices)>1:
        valid_values = site_v_d[valid_indices]
        interpolator = interp1d(valid_indices, valid_values, kind='nearest', fill_value='extrapolate')
        ndvi_16day_imp = interpolator(numpy.arange(len(site_v_d)))
    else:
        print('all are nan.')
        ndvi_16day_imp = site_v_d[:]
    # 16daily to 10-min
    # numpy.arange(1, 366, 16)[-1:][0]+15-365 --> 3
    site_v_10min_o = numpy.repeat(ndvi_16day_imp, 16*6*24)
    site_v_2018 = site_v_10min_o[:(365*6*24)]
    site_v_2019 = site_v_10min_o[(365*6*24):(2*365*6*24)]
    site_v_10min = numpy.concatenate((site_v_2018, site_v_2019))
    site_record_10min = []
    current_date = data_first_date
    for site_v in site_v_10min:
        current_day_time = current_date.strftime("%Y-%m-%dT%H:%M:%SZ")
        site_record_10min.append([current_day_time, str(site_v)])
        current_date = current_date + timedelta(minutes=time_internal)
    ndvi_record = numpy.array(site_record_10min)
    save_filename = os.path.join(WORKSPACE_FOLDER, site_infos[site_idx][0]+'_MOD13A2_NDVI'+'.csv')
    print(save_filename)
    numpy.savetxt(save_filename, ndvi_record, delimiter=",", fmt='%s')
```
