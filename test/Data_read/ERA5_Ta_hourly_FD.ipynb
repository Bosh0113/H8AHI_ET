```python
# FD extent
resolution = 0.01 # degree
lats = numpy.arange(60.-resolution/2, -60, -resolution)
lons = numpy.arange(85.+resolution/2, 180, resolution)
```

```python
day_num = 1 # 1 -> 31(?)
hour_idx = 2 # 0 <=> 01:00:00

ta_dh = ta_ds.isel(time=day_num, step=hour_idx)
fd_ds = ta_dh.interp(longitude=lons, latitude=lats, method="linear", kwargs={"fill_value": "extrapolate"}) # linear
ta_v = fd_ds.to_array()[0]-273.15
```

```python
plt.figure(figsize=(10,10))
plt.imshow(ta_v, interpolation='None')
plt.colorbar(label='℃')
plt.title('0.01° Ta in study extent at 2018-01-01 03:00:00')
```
