```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
def read_site_ref(csv_filename):
    csv_data = numpy.genfromtxt(csv_filename, delimiter=',', dtype=str)
    data_10min = csv_data[:, 1]
    data_10min = data_10min.astype(float)
    return data_10min
```

```python
import matplotlib.pyplot as plt
import numpy as np

site_names = ['TKY', 'CRK', 'GCK', 'GDK']

fig, axes = plt.subplots(len(site_names), 1, figsize=(10, 4 * len(site_names)))

for i, site_name in enumerate(site_names):
    Rsd_SITE_CSV = '/disk2/Data/FluxSite4ET/' + site_name + '_AMATERASS_Rsd.csv'
    data_10min = read_site_ref(Rsd_SITE_CSV)
    data_day = data_10min.reshape(365 * 2, 24 * 6)

    axes[i].plot(np.arange(365 * 2), data_day)
    axes[i].set_title(site_name + ' 2018-2019')
    axes[i].set_xlim(0, 365 * 2)
    axes[i].set_ylabel('AMATERASS Rsd')

plt.tight_layout()
plt.show()
```
