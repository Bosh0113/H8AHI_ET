```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
def read_site_wv(csv_filename):
    csv_data = numpy.genfromtxt(csv_filename, delimiter=',', dtype=str)
    wv_day = csv_data[:, 1]
    wv_day = wv_day.astype(float)
    return wv_day
```

```python
MOD05_TPWV_SITE_CSV = '/disk2/Data/FluxSite4ET/original/TKY_MOD05_TPWV.csv'
MYD05_TPWV_SITE_CSV = '/disk2/Data/FluxSite4ET/original/TKY_MYD05_TPWV.csv'
mod05_wv_day = read_site_wv(MOD05_TPWV_SITE_CSV)
myd05_wv_day = read_site_wv(MYD05_TPWV_SITE_CSV)
plt.figure(figsize=(10,4))
plt.plot(numpy.arange(365*2), mod05_wv_day, color='red', label='MOD05 TPWV', alpha=0.6)
plt.plot(numpy.arange(365*2), myd05_wv_day, color='green', label='MYD05 TPWV', alpha=0.6)
plt.title('TKY 2018-2019 MOD05|MYD05 Total Precipitable Water')
plt.xlim(0, 365*2)
plt.ylabel('Atmospheric total column water vapor (cm)')
```
