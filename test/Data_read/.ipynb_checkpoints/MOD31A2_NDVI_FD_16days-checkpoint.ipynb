## Read and clip data to FD extent

```python
import os
import xarray
import numpy
import matplotlib.pyplot as plt
```

```python
MOD31A2_NDVI_FOLDER = '/remote_geo01/data_fd_et/MOD31A2_NDVI'
```

```python
# FD extent
resolution = 0.01 # degree
lats = numpy.arange(60.-resolution/2, -60, -resolution)
lons = numpy.arange(85.+resolution/2, 180, resolution)
```

```python
mod11_filename = os.path.join(MOD31A2_NDVI_FOLDER, 'MOD13A2.061__1_km_16_days_NDVI_doy2018001_aid0001.tif')
modis_ds = xarray.open_rasterio(mod11_filename)[0]
fd_ds = modis_ds.interp(x=lons, y=lats, method="nearest", kwargs={"fill_value": "extrapolate"}) # linear
data_dn = numpy.array(fd_ds.values)
data_dn[data_dn==-3000] = numpy.NaN
data_v = data_dn*0.0001
```

```python
print(mod11_filename)
data_show = numpy.copy(data_v)

fig, ax = plt.subplots(figsize=(10, 10))
plt.imshow(data_show, interpolation='None')
plt.colorbar()
```
