```python
import numpy
import matplotlib.pyplot as plt
```

```python
def calculate_r_net(r_sd, r_su, r_ld, r_lu):
    r_net = (r_sd - r_su) + (r_ld - r_lu)
    return r_net
```

```python
SITE_CSV = '/disk2/Data/FluxSite4ET/validation/CRK_2018_2019_L2.csv'
csv_data = numpy.genfromtxt(SITE_CSV, delimiter=',', dtype=str)
csv_data[0].tolist()
```

```python
data_labels = ['Rsd', 'Rsu', 'Rld', 'Rlu']
fig, axes = plt.subplots(len(data_labels)+1, 1, figsize=(10, 4 * (len(data_labels)+1)))
for idx in range(len(data_labels)):
    r_data = csv_data[:,idx+1][1:].astype(float)
    data_day = r_data.reshape(365 * 2, 24 * 2)
    axes[idx].plot(numpy.arange(365 * 2), data_day)
    axes[idx].set_title('CRK 2018-2019')
    axes[idx].set_xlim(0, 365 * 2)
    axes[idx].set_ylabel('CRK Radiation ' + data_labels[idx])

r_net_data = calculate_r_net(csv_data[:,1][1:].astype(float), csv_data[:,2][1:].astype(float), csv_data[:,3][1:].astype(float), csv_data[:,4][1:].astype(float))
data_day = r_net_data.reshape(365 * 2, 24 * 2)
axes[4].plot(numpy.arange(365 * 2), data_day)
axes[4].set_title('CRK 2018-2019')
axes[4].set_xlim(0, 365 * 2)
axes[4].set_ylabel('CRK Net Radiation')
    
plt.tight_layout()
plt.show()
```
