```python
def combine_and_mask_data(SITE_NAME):
    
    def load_data(year):
        """Load data from the specified year."""
        filename = f'/geoland01/share/people/liwei/Asiaflux_2023/cloudmask/{SITE_NAME}_{year}.h5'
        with h5py.File(filename, 'r') as hf:
            # List the content and read datasets
            # print(f'List of arrays in {year} file: \n', list(hf.keys()))
            
            data = {}
            for key in hf.keys():
                data[key] = np.array(hf.get(key))
        return data

    # Load 2018 and 2019 data
    data_2018 = load_data(2018)
    data_2019 = load_data(2019)
    
    # Check the shapes for compatibility
    for key in data_2018.keys():
        assert data_2018[key].shape == data_2019[key].shape, f"Mismatched shape for {key}: {data_2018[key].shape} vs {data_2019[key].shape}"

    # Combine the datasets
    combined_data = {}
    for key in data_2018.keys():
        combined_data[key] = np.concatenate((data_2018[key], data_2019[key]), axis=0)

    # Apply masks
    def apply_mask(data_array, mask_array, condition_value=1):
        """Apply mask based on a condition."""
        # Check shape compatibility for masking
        assert data_array.shape == mask_array.shape, f"Mismatched shapes: {data_array.shape} vs {mask_array.shape}"
        return np.where(mask_array >= condition_value, data_array, np.nan).reshape(730, 54)
    
    masks = {}
    keys_to_mask = ['B1', 'B2', 'B3', 'B4', 'B5', 'SAA', 'SZA', 'VAA', 'VZA']

    for key in keys_to_mask:
        data = combined_data[key][:, 1, 1]
        if key in ['B1', 'B2', 'B3', 'B4', 'B5']:
            data = np.where(data == 3.2768, np.nan, data)
            data = np.where(data <= -1, np.nan, data)
        
        masks[f"{key}_mask"] = apply_mask(data, combined_data['CloudMask'][:, 1, 1])

    return masks


def outlier_detection_fill_nan(data, threshold=2):
    """
    使用三倍标准差进行异常值检测，并将异常值填充为 NaN
    
    参数:
        data: 一个包含数值数据的一维NumPy数组或类似的数据结构
        threshold: 异常值的阈值，默认为3
        
    返回值:
        一个新的NumPy数组，将异常值填充为 NaN
    """
    data = data.copy()  # 复制数据以保持原始数据不受影响
    
    mean = np.nanmean(data)
    std = np.nanstd(data)
    
    lower_bound = mean - threshold * std
    upper_bound = mean + threshold * std
    
    outliers = (data < lower_bound) | (data > upper_bound)
    
    data[outliers] = np.nan
    
    return data

from scipy.signal import savgol_filter
from scipy.interpolate import interp1d

def nearest_neighbor_interpolation(data):
    """Fill NaN values using nearest neighbor interpolation."""
    n = len(data)
    
    # Indices of NaN values
    nan_indices = np.where(np.isnan(data))[0]
    
    # Indices of non-NaN values
    valid_indices = np.where(~np.isnan(data))[0]
    
    # For each NaN, find the index of the nearest valid value
    nearest_indices = np.searchsorted(valid_indices, nan_indices)
    
    # Correct indices that are out of bounds
    nearest_indices[nearest_indices == len(valid_indices)] -= 1
    
    # Get the nearest valid values
    nearest_values = data[valid_indices[nearest_indices]]
    
    # Fill NaN values with the nearest valid values
    data[nan_indices] = nearest_values
    
    return data


def smooth_data(data, window_size=16, polynomial_order=1):
    
    # Find indices of non-NaN values
    non_nan_indices = np.where(~np.isnan(data))[0]
    non_nan_values = data[non_nan_indices]

    # Interpolate to fill NaN values
    f = interp1d(non_nan_indices, non_nan_values, kind='linear', fill_value='extrapolate')
    interpolated_data = f(np.arange(len(data)))
    # interpolated_data = nearest_neighbor_interpolation(data)
    # Apply Savitzky-Golay filter
    smoothed_data = savgol_filter(interpolated_data, window_size, polynomial_order)
    
    return smoothed_data

def RTK (sza,vza,raa):
    cos_xi = np.cos(np.radians(sza)) * np.cos(np.radians(vza)) + np.sin(np.radians(sza)) * np.sin(np.radians(vza))* np.cos(np.radians(raa))
    xi = np.arccos(cos_xi)
    rtk = (((np.pi/2 - xi) * cos_xi + np.sin(xi)) / (np.cos(np.radians(sza)) + np.cos(np.radians(vza)))) - np.pi/4
    return rtk
def LSR_MaxMin(sza,vza,raa):
    cos_sza = np.cos(np.radians(sza))
    cos_vza = np.cos(np.radians(vza))
    cos_raa = np.cos(np.radians(raa))

    sin_sza = np.sin(np.radians(sza))
    sin_vza = np.sin(np.radians(vza))
    sin_raa = np.sin(np.radians(raa))

    tan_sza = np.tan(np.radians(sza))
    tan_vza = np.tan(np.radians(vza))

    sec_sza = 1 / cos_sza
    sec_vza = 1 / cos_vza

    _sza = np.arctan(tan_sza)
    _vza = np.arctan(tan_vza)

    cos__sza = np.cos(_sza)
    cos__vza = np.cos(_vza)
    sin__sza = np.sin(_sza)
    sin__vza = np.sin(_vza)
    sec__sza = 1 / cos__sza
    sec__vza = 1 / cos__vza

    cos__xi = cos__sza * cos__vza + sin__sza * sin__vza * cos_raa
    D = np.sqrt(np.square(tan_sza) + np.square(tan_vza) - 2 * tan_sza * tan_vza * cos_raa)
    cos_t = (2 * (np.sqrt(np.square(D) + np.square(tan_sza * tan_vza * sin_raa)))) / (sec__sza + sec__vza)
    
    cos_t = np.where(cos_t>1,1,cos_t)
    cos_t = np.where(cos_t<-1,-1,cos_t)
    
    
    t = np.arccos(cos_t)
    O = (1/np.pi) * (t - np.sin(t) * cos_t) *  (sec__sza + sec__vza)
    lsr = O - sec__sza - sec__vza + 0.5 * (1 + cos__xi) * sec__sza * sec__vza
    return lsr

from sklearn.linear_model import LinearRegression
def fitting(ref,dg,dv):
    ref = np.array(ref)
    dg = np.array(dg)
    dv = np.array(dv)
    
    r = ref.reshape(-1,1)
    cm = ~np.isnan(r)


    r = r.reshape(-1,1)[cm].reshape(-1,1)
    kv = dv.reshape(-1,1)[cm].reshape(-1,1)
    kg = dg.reshape(-1,1)[cm].reshape(-1,1)

    cm = ~np.isnan(kv)
    r = r[cm].reshape(-1,1)
    kv = kv.reshape(-1,1)[cm].reshape(-1,1)
    kg = kg.reshape(-1,1)[cm].reshape(-1,1)

    cm = ~np.isnan(kg)
    r = r[cm].reshape(-1,1)
    kv = kv.reshape(-1,1)[cm].reshape(-1,1)
    kg = kg.reshape(-1,1)[cm].reshape(-1,1)
    
    # 检查是否存在空值，并去除
    if not (np.isnan(kg) | np.isnan(kv)).any():
        k = np.hstack((kv, kg))
    else:
        
        return np.nan, np.nan, np.nan
    
    # k = k.reshape(1,-1)
    # r = r.reshape(1,-1)
    # 检查数据量是否大于40
    if r.size > 20:
        linreg = LinearRegression()
        model = linreg.fit(k, r)
        fi = linreg.intercept_[0]
        fv, fg = linreg.coef_[0]
        return fi, fv, fg
    else:
        return np.nan, np.nan, np.nan
    
# Get Label
import datetime as dt
def Time_split_AHI(time):
    date = time.strftime("%Y%m%d%H%M")
    YYYY, MM, DD, HH, MIN = date[0:4], date[4:6], date[6:8], date[8:10], date[10:12]
    return YYYY,MM,DD,HH,MIN,date

import ephem
# from datetime import dt

def solar_noon(date, observer_latitude, observer_longitude):
    """
    Calculate solar noon using the ephem library.

    :param date: Date for which to calculate solar noon (e.g., '2023-10-18')
    :param observer_latitude: Latitude of the observer (e.g., '37.7749') 
    :param observer_longitude: Longitude of the observer (e.g., '-122.4194')

    :return: Solar noon time as an ephem.Date
    """
    observer = ephem.Observer()
    observer.date = date
    observer.lat = observer_latitude
    observer.lon = observer_longitude

    sun = ephem.Sun(observer)
    noon = observer.next_transit(sun)
    formatted_str = noon.datetime().strftime('%Y-%m-%d %H:%M')

# 将格式化的字符串解析为 datetime 对象
    new_dt = dt.datetime.strptime(formatted_str, '%Y-%m-%d %H:%M')
    # noon = datetime.fromtimestamp(noon)
    
    return new_dt


def calc_sunpos(dtime,lat,lon):
    sun = ephem.Sun()
    obs = ephem.Observer()
    obs.date = dtime
    obs.lat = lat
    obs.long = lon
    sun.compute(obs)
    return np.degrees(sun.az),90.0-np.degrees(sun.alt)


# date_start = '2018-01-01 01:00'
# date_end = '2018-01-01 07:00'
# date_t = dt.timedelta(minutes=10)
# # date_t = dt.timedelta(minutes=30)
# date_s = dt.datetime.strptime(date_start, "%Y-%m-%d %H:%M")
# date_e = dt.datetime.strptime(date_end, "%Y-%m-%d %H:%M")
# date_time_now = date_s
# label = []
# while date_time_now < date_e:
#     # print(date_time_now)
#     YYYY,MM,DD,HH,MIN,date= Time_split_AHI(date_time_now)
#     label.append(HH+MIN)
#     date_time_now = date_time_now + date_t
```

```python

# Example usage:
def rolling_mean_corrected(data_array, window_size,value):
    # Padding the original data with nan values
    half_window = window_size // 2
    padded_data = np.concatenate(([np.nan] * half_window, data_array, [np.nan] * half_window))

    # Calculate the rolling mean on the padded data
    rolling_means = []
    for i in range(len(data_array)):
        window = padded_data[i:i+window_size]
        mean = np.nanmedian(window)
        rolling_means.append(mean)

    # Convert rolling_means to numpy array
    rolling_means = np.array(rolling_means)

    # Calculate the mask
    mask = (data_array - rolling_means) / rolling_means

    # Update data_array based on the mask
    data_array = np.where(mask < -value, np.nan, data_array)
    # data_array = np.where(mask > value, np.nan, data_array)
    return data_array


SR_EVI2 = 2.5 * (B4_mask - B3_mask) / (B4_mask + 2.4 * B3_mask + 1)
# SR_EVI2 = (B4_mask - B3_mask) / (B4_mask + B3_mask)

for i in range(730):
    SR_EVI2[i,0:54] = outlier_detection_fill_nan(SR_EVI2[i,0:54], threshold=1)

SR_EVI2_MAX = np.nanmax(SR_EVI2[:,0:54],axis=1)
SR_EVI2_MAX = np.where(SR_EVI2_MAX<0,np.nan,SR_EVI2_MAX)
SR_EVI2_MAX = rolling_mean_corrected(SR_EVI2_MAX, 30,0.15)
SR_EVI2_MAX = rolling_mean_corrected(SR_EVI2_MAX, 7,0.05)
SR_EVI2_MAX = rolling_mean_corrected(SR_EVI2_MAX, 7,0.05)

smoothed_SR = smooth_data(SR_EVI2_MAX,window_size=12, polynomial_order=1)

SR_EVI2_MAX1 = np.nanmax(SR_EVI2[:,0:54],axis=1)
```
